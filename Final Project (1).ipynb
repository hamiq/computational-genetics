    sequence should be trimmed.

using the gzip module and the SeqIO module from Biopython. Within the
loop, I counted the number of sequences there are in the file by
accumulating the number of loops that ran and to find the total number
of basepairs, I accumulated the length of each sequence in the file.
Next, I generated a dictionary of the quality scores for each nucleotide
in each sequence and used indexing to work with the first four and last
four nucleotides in each sequence. I added the scores of the first four
together and, separately, the last four together and divided by four to
get the averages in each sequence. Next, I took the sum of all of the
beginning average quality scores and, again separately, the sum of all
of the final average quality scores to calculate the overall average
scores for the first four and last four nucleotides of the file.
Finally, I averaged the quality scores of every nucleotide in each
sequence and summed up those values to calculate the overall quality
score of the entire file.

    import gzip
    from Bio import SeqIO

    original_seq=0              # number of sequences
    original_bp=0               # number of basepairs
    original_seq_qual=0         # average quality score for entire sequence
    sum_first_avg=0             # sum of the averages of the first 4 bases of every sequence
    sum_last_avg=0              # sum of the averages of the last 4 bases of every sequence
    sum_avg=0                   # sum of averages of each sequence
    original_first_avg=0        # average quality score for first 4 bases
    original_last_avg=0         # average quality score for last 4 bases

    fh=gzip.open('hi70.unaligned.fastq.gz', 'rt')

    for record in SeqIO.parse(fh, 'fastq'):
        original_seq+=1
        num_bp=len(record)
        original_bp+=num_bp
        qlist=record.letter_annotations['phred_quality']
        sum_first=0
        for i in range(4):
            sum_first+=int(qlist[i])
        avg_first_seq=sum_first/4
        sum_first_avg+=avg_first_seq
        sum_last=0
        for i in range(-4,0):
            sum_last+=int(qlist[i])
        avg_last_seq=sum_last/4
        sum_last_avg+=avg_last_seq
        seq_qual=0
        for value in qlist:
            seq_qual+=int(value)
        average=seq_qual/len(record)
        sum_avg+=average

    original_first_avg=sum_first_avg/original_seq
    original_last_avg=sum_last_avg/original_seq
    original_seq_qual=sum_avg/original_seq

    print(original_seq)
    print(original_bp)
    print(original_first_avg)
    print(original_last_avg)
    print(original_seq_qual)
